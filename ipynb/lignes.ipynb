```python
import requests
import json
from bs4 import BeautifulSoup
```

Extraction des Arrêts
=====================

```python
URL = "http://www.lestis72.com/horaires_ligne/index.asp?rub_code=6&date=19%2F09%2F2014&lign_id="
```

```python
with open('parsed.json') as f:
    liste_lignes = json.loads(f.read())
```

```python
communes = []
for idx in liste_lignes.keys():
    print(idx)
    url = URL+str(idx)
    req = requests.get(url)
    req.encoding = "ISO-8859-1"
    soup = BeautifulSoup(req.content)
    communes += map(lambda _: _.contents, soup.find(id='linehour').findAll(headers="commune"))
    
    
```

```python
communes = list(set(map(lambda _: _[0], communes)))
```

Passage en conformité INSEE
===========================

```python
import re

lelales = re.compile(r'^(LE|LA|LES)[\s-]')
sur = re.compile(r'\s?/\s?')
sous = re.compile(r'\s?\\\s?')
st = re.compile(r'(-|^)ST-')
ste = re.compile(r'(-|^)STE-')

exceptions = {
    "MALICORNE": "MALICORNE-SUR-SARTHE",
}

def inseeify(string):
    string = lelales.sub('', string)
    string = sous.sub(' SOUS ', string)
    string = sur.sub(' SUR ', string)
    string = string.replace(' ', '-')
    string = st.sub('-SAINT-', string)
    string = ste.sub('-SAINTE-', string)
    string = string.replace('É', 'E')
    if string[0]=='-': string = string[1:]
        
    if string in exceptions.keys():
        string = exceptions[string]
        
    return string
    
```

```python
comunes_insee = list(map(inseeify, communes))
```

```python
insee_hashtable = {}

from codecs import open as copen

with copen('Code_insee.txt', encoding="utf-16") as f:
    line = f.readline()
    
    while line:
        line = line.rstrip()
        splitted = line.split('\t')
        insee_hashtable[splitted[0]] = splitted[1]
        line = f.readline()
```

Jeu de test: Ligne 4
====================

```python
bus_stops = []

with copen('ARRET_L04.txt', encoding="utf-16") as f:
    line = f.readline()
    
    while line:
        line = line.rstrip()
        bus_stops.append(tuple(line.split('\t')))
        line = f.readline()
    
```

```python
with open('tis2014_mercator.json') as f:
    cg_stops = json.loads(f.read())
```

```python
stops_coords = {}

for s in cg_stops['features']:
    if not stops_coords.get(s['properties']['COD_COMM']):
        stops_coords[s['properties']['COD_COMM']] = {}
        
    stops_coords[s['properties']['COD_COMM']][s['properties']['NOM']] = s['geometry']['coordinates']
```

```python
out_json = {"type": "FeatureCollection", "features": []}

ligne = "4"

for commune,arret in bus_stops:
    if stops_coords.get(insee_hashtable[inseeify(commune)]):
        try:
            coords = stops_coords.get(insee_hashtable[inseeify(commune)])[arret]
        except KeyError as e:
            print(e)
            print(insee_hashtable[inseeify(commune)])

        out_json['features'].append({
            "type": "Feature",
            "geometry": {
                "type": "Point",
                "coordinates": coords
            },
            "properties": {
                "Ligne TIS": ligne,
                "Nom de l'arret": commune+" - "+arret,
            }
        })
```

```python
with open('ligne'+ligne+'_test.json', 'w') as f:
    f.write(json.dumps(out_json))
```
